- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

iris = datasets.load_iris()
X = iris.data
Y= iris.target

```

```python
estimators = [('k_means_5', KMeans(n_clusters=3)),
              ('k_means_9', KMeans(n_clusters=8)),
              ('k_means_bad_init', KMeans(n_clusters=3, n_init=1, init='random'))]
```

```python
fignum = 1
titles = ['3 clusters', '9 clusters', '3 clusters with bad init']

for name, est in estimators:
    fig = plt.figure(num = fignum, figsize = (6 , 4))
    ax = Axes3D(fig = fig)
    
    est.fit(X)
    labels = est.labels_
    
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c= labels)
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum-1])
    fignum +=1
```

```python
fig = plt.figure(num = fignum, figsize = (4,3))
ax = Axes3D(fig = fig)

for name, label in zip(iris.target_names, [0,1,2]):
    ax.text3D(X[Y==label, 0].mean(),
              X[Y==label, 1].mean(),
              X[Y==label, 2].mean(), name,
              horizontalalignment = 'center')

ax.scatter(X[:,0], X[:,1], X[:,2], c=Y)
```
